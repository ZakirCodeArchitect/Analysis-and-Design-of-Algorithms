```python
import matplotlib.pyplot as plt

# Data
n_values = [0, 5, 10, 15, 20, 25, 30, 35, 40]
iterative_times = [1012, 8446, 11089, 15002, 33043, 21887, 30330, 37685, 17041]  # in milliseconds
recursive_times = [1060, 1829, 7138, 1062, 19799, 30399, 50091, 197062, 1582373]  # in milliseconds

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, iterative_times, marker='o', label='Iterative Approach')
plt.plot(n_values, recursive_times, marker='o', label='Recursive Approach')
plt.title('Execution Time of Fibonacci Series Generation')
plt.xlabel('Value of n')
plt.ylabel('Execution Time (milliseconds)')
plt.xticks(n_values)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
